## 2.2.1: Reading and Writing Data

Файл [grades.csv](grades.csv) представляет собой файл с именами учащихся и буквенными оценками:

```
Alice; A
Bob; B
Robert; A
Richard; C
```

Записать данные из файла в массив. Добавьте средний балл учащегося в строку учащегося (A=4,B=3,C=2,D=1).

Подсказка: функция [strip()](https://docs.python.org/3/library/stdtypes.html#str.strip) удаляет конечные пробелы из строки.

Запишите этот файл в новый файл `grades_gpa.csv`.

````python
import pandas as pd
import numpy as np
````

````python
from google.colab import files                  #загрузка данных в колаб
uploaded = files.upload()

for fn in uploaded.keys():
  print('User upload file "{name}" witn length {length} bytes'.format(name=fn,length=len(uploaded[fn])))
````

````python
#Запись данных сналача в датафрейм (мне так было удобнее):
grades_df=pd.read_csv("grades.csv")
grades_df=pd.concat([pd.DataFrame([grades_df.columns.values], columns=grades_df.columns), grades_df], ignore_index=True) #настраиваем, чтобы первая строка была строкой, а не заголовками


#Разделяем столбец с именем и оценкой на два:
grades_df[["Name", "Grade"]] = pd.DataFrame(grades_df["Alice; A"].str.split("; ",expand=True), index = grades_df.index)
del grades_df['Alice; A']
grades_df
````

````python
#Из исходного словаря сделаем датафрейм:
gpas = {"A":4, "B":3, "C":2, "D":1}
gpas_df=pd.DataFrame(list(gpas.items()), columns=['Grade',"Number"])

#Теперь объеденим два датафрейма:
grades_gpas_df=pd.merge(grades_df,gpas_df,on='Grade',how='inner')
grades_gpas_df
````

````python
#Запишем полученный датафрейм в csv файл:
grades_gpas_df.to_csv('grades_gpa.csv', index=False)

print("Массив успешно записан в CSV файл 'grades_gpa.csv'")
````


 * Создайте новый фрейм данных на основе набора данных hit_albums, который содержит только исполнителей, у которых в списке есть как минимум два альбома.

````python
#Создадимновый датафрейм, в котором будут посчитано количество альбомов у каждого артиста
df2=hit_albums
df2=df2.groupby(['Artist'])['Album'].count().reset_index(name='Amount_of_albums')
df2=df2.loc[(df2['Amount_of_albums']>=2)]

#Объединим два датафрейма, чтобы теперь выводились только те артисты, у которых альбомов выпущено больше 2
hit_albums_2=pd.merge(hit_albums,df2,on='Artist',how='inner')
hit_albums_2.sort_values('Artist')
````
